#### Логічні оператори `and`, `or`, `not`

Ними комбінують кілька умов — саме так будуються складні фільтри даних:
- `and` повертає `True`, тільки якщо **всі** умови істинні
- `or` повертає `True`, якщо **хоча б одна** умова істинна
- `not` — заперечення (інвертує значення)

```python
revenue = 120000
orders = 350

# VIP-клієнт: багато витратив І багато замовлень
is_vip = revenue > 100000 and orders > 300
print(is_vip)   # True
```

```python
# потрібна увага: або мало замовлень, АБО мала виручка
needs_attention = orders < 100 or revenue < 50000
print(needs_attention)   # False
```

#### Корисні методи рядків

Метод — це дія, яку можна виконати над значенням через крапку. Ось найпоширеніші для тексту:

- `s.lower()` — переводить усі символи в нижній регістр
- `s.upper()` — переводить усі символи у верхній регістр
- `s.strip()` — прибирає пробіли на початку та в кінці
- `s.replace(old, new)` — замінює один фрагмент іншим
- `s.split(separator)` — розбиває рядок на список за роздільником
- `s.startswith(...)` / `s.endswith(...)` — перевіряє початок/кінець

Це надзвичайно корисно при очищенні «брудних» текстових даних (зайві пробіли, різний регістр).

```python
raw_city = "  київ  "

print(raw_city.strip())            # прибрали пробіли -> "київ"
print(raw_city.strip().title())    # перша літера велика -> "Київ"
print(len(raw_city))               # довжина рядка (з пробілами)
print(len(raw_city.strip().title()))
```

```python
email = "IVAN.KOVALENKO@GMAIL.COM"
print(email.lower())               # нормалізуємо до нижнього регістру
print(email.endswith("gmail.com".upper()))  # перевірка домену
print(email.endswith("meta.ua".upper()))  # перевірка домену
```

Для неіменованих аргументів працює наступним чином:

![](https://drive.google.com/uc?export=view&id=1EYiFLfByrc42mEHc4Nk41M3aZ6-kWSVS)

Форматувати числа з плаваючою комою можемо наступним чином:

![](https://drive.google.com/uc?export=view&id=1fjQE8DimPSWFFyCeFz2LmN1gPVlGxEo3)

```python
# аргументи за замовченням
print("Hello {}, your balance is {}.".format("Adam", 230.2346))

# позиційні аргументи
print("Hello {0}, your balance is {1}.".format("Adam", 230.2346))

# іменовані аргументи
print("Hello {name}, your balance is {blc}.".format(name="Adam", blc=230.2346))

# змішані аршументи
print("Hello {0}, your balance is {blc}.".format("Adam", blc=230.2346))
```

Форматування цифр:

```python
# integer arguments
print("The number is: {:d}".format(123))

# float arguments
print("The float number is: {:.2f}".format(123.4567898))

# octal, binary and hexadecimal format
print("bin: {0:b}, oct: {0:o}, hex: {0:x}".format(12))
```

Можемо до форматування додати паддінг:

```python
# мінімальна ширина вікна для числа
print("{:5d}".format(12))

# встановлення ширини вікна не спрацює, якщо передане число більше за встановлене обмеження
print("{:2d}".format(1234))

# падінг для чисел з плаваючою комою
print("{:8.3f}".format(12.2346))

# заповенння нулями для цілих чисел
print("{:05d}".format(12))
print("{:05d}".format(1112))
print("{:05d}".format(1))

# заповенння нулями для чисел з плаваючою комою
print("{:08.3f}".format(12.2346))
```

#### Корисні функції для списків чисел

### Словники (Dictionaries) — зберігання пар «ключ–значення»

**Словник** зберігає дані у вигляді пар `ключ: значення`. Це найзручніший спосіб описати один об'єкт з іменованими полями — наприклад, одного клієнта чи один рядок таблиці. Саме у форматі словника приходять дані з веб-API (JSON).

```python
# Один клієнт як словник
customer = {
    "name": "Іван Коваленко",
    "city": "Київ",
    "orders": 12,
    "total_spent": 45300.50,
    "is_active": True,
}
customer
```

#### Доступ до значень за ключем

```python
print(customer["name"])                  # Іван Коваленко
print(customer.get("orders"))            # 12
print(customer.get("phone", "не вказано"))  # ключа немає -> значення за замовчуванням
```

Зверніть увагу на різницю: `customer["phone"]` викине помилку, якщо ключа немає, а `customer.get("phone", ...)` поверне значення за замовчуванням. У реальних даних із пропусками `.get()` часто безпечніший.

#### Зміна та додавання даних

```python
customer["orders"] = 13              # оновили значення
customer["last_order"] = "2026-07-10"  # додали новий ключ
customer
```

#### Методи словника: ключі, значення, пари

```python
print(customer.keys())     # усі ключі (назви полів)
print(customer.values())   # усі значення
print(customer.items())    # пари ключ-значення
```

### Множини (Sets) — унікальні елементи

**Множина** — це колекція **унікальних** елементів без порядку. Автоматично видаляє дублікати. Незамінна річ, коли треба відповісти на питання «скільки унікальних клієнтів / міст / категорій?».

```python
# У даних місто повторюється багато разів
cities_raw = ["Київ", "Львів", "Київ", "Одеса", "Львів", "Київ"]

unique_cities = set(cities_raw)   # тільки унікальні
print(unique_cities)
print("Унікальних міст:", len(unique_cities))
```

#### Операції над множинами

Над множинами працюють математичні операції — об'єднання, перетин, різниця:

![](https://media.geeksforgeeks.org/wp-content/uploads/20240619161350/Set-Operations.webp)

```python
buyers_january = {"Іван", "Марія", "Олег", "Анна"}
buyers_february = {"Марія", "Олег", "Петро"}

# Хто купував в обидва місяці? (перетин)
print("Обидва місяці:", buyers_january & buyers_february)

# Усі, хто купував хоч колись? (об'єднання)
print("Усього клієнтів:", buyers_january | buyers_february)

# Хто був у січні, але зник у лютому? (різниця)
print("Відпали:", buyers_january - buyers_february)
```

## Дебаг коду — як знаходити помилки

Ніхто не пише код без помилок з першого разу. **Дебаг** (debugging) — це процес пошуку та виправлення помилок. Уміння читати помилки й розуміти, що пішло не так — навичка, яка економить години. Розгляньмо кілька способів, від найпростішого.

### Спосіб 1: Виводити проміжні значення через `print`

Найпростіший спосіб зрозуміти, що відбувається всередині коду — додавати `print()` у ключових місцях. Часто помилка в тому, що дані виявилися не такими, як ви очікували.

Ось приклад: чому підсумковий список порожній?

```python
# Приклад: чомусь результат порожній
data = [41, 12, 11, 60, 1, 45, 67, 23, 89]
threshold = 200
big_values = []

for value in data:
    if value > threshold:   # поріг завеликий!
        big_values.append(value)

print(big_values)   # []
```

Додамо `print`, щоб побачити, що відбувається на кожній ітерації:

```python
data = [41, 12, 11, 60, 1, 45, 67, 23, 89]
threshold = 200
big_values = []

print(f"DEBUG: поріг = {threshold}")
for value in data:
    print(f"DEBUG: перевіряю {value}, {value} > {threshold} = {value > threshold}")
    if value > threshold:
        big_values.append(value)

print(f"DEBUG: фінальний список = {big_values}")
```

Одразу видно: жодне значення не проходить умову, бо поріг `200` завеликий. Виправимо на реальний поріг:

```python
data = [41, 12, 11, 60, 1, 45, 67, 23, 89]
threshold = 50
big_values = []

for value in data:
    if value > threshold:   # тепер поріг адекватний
        big_values.append(value)

print(f"Результат: {big_values}")   # [60, 67, 89]
```

`print`-дебаг корисний і в функціях — щоб бачити, які значення й типи туди приходять:

```python
def process_value(value):
    print(f"DEBUG: value = {value}, тип = {type(value)}")

    if isinstance(value, str):
        print("DEBUG: це рядок, конвертуємо в число")
        value = int(value)

    result = value * 2
    print(f"DEBUG: результат = {result}")
    return result

process_value("5")   # прийшов текст
process_value(10)    # прийшло число
```

### Спосіб 2: Читати traceback (текст помилки)

Коли код падає з помилкою, Python виводить **traceback** — «слід» помилки. Головне правило: **читати його знизу вгору**. Найнижчий рядок каже, *що саме* пішло не так, а рядки вище показують, *де* це сталося.

Запустіть комірку нижче й подивіться на помилку:

```python
# Код з помилкою
def divide(a, b):
    return a / b

numbers = [10, 20, 30, 0, 40]   # тут є 0!
results = []

for num in numbers:
    result = divide(100, num)   # ділення на 0 при num=0
    results.append(result)
```

В останньому рядку traceback ви побачите `ZeroDivisionError: division by zero`. Це і є суть проблеми: десь відбулося ділення на нуль. Рядки вище підказують, що це сталося у функції `divide` при обробці значення `0` зі списку.

**Найчастіші помилки, які ви бачитимете:**
- `NameError` — звернулися до змінної, якої не існує (частий кейс: не запустили комірку, де вона створюється)
- `TypeError` — операція над несумісними типами (наприклад, `"5" + 5`)
- `KeyError` — звернулися до ключа, якого немає у словнику
- `ValueError` — правильний тип, але недопустиме значення (`int("abc")`)
- `ZeroDivisionError` — ділення на нуль
- `IndexError` — звернулися до неіснуючого індексу списку
